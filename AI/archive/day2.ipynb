```python
def get_abbr(x):
    city_abbr = []
    for i in x:
        city_abbr.append(i[0:3])
    return city_abbr

city_list = ['Seoul', 'Daegu', 'Kwangju', 'Jeju']
print(get_abbr(city_list))
```
