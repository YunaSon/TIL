함수 밖의 정의는 함수 안으로 가져올 수 있지만 
함수 안의 정의는 함수 밖에서 사용할 수 없다 

정의를 가져오는 우선순위 LEGB   local 함수 안의 정의 enclosing  
global 작업하고 있는 곳에서 bounding 함수 안에서는 접근을 할 수 있지만 함수 안에서 함수 밖의 정의를 변경할 수 없다
