```python
#1. 미분하고자 하는 함수를 정의한다. 
def func1(x):
    return x**2
```

```python
# 2. x가 호출되면, func1(x)를 부르겠다.
f = lambda x : func1(x)
```

```python
def numerical_derivative(f, x):
    delta_x = 1e-4
    return (f(x+delta_x) - f(x-delta_x))/ (2*delta_x)
```

```python
ret1 = numerical_derivative(f, 3.0) #람다에 의해 정의된 f (가독성 좋다. C언어에서는 함수 포인터를 이용한다.)
print(ret1)
ret2 = numerical_derivative(func1, 3.0) #func1 (3.0이 미세하게 변할 때 func1이 얼마나 변하는가)
print(ret2)
```

```python
import numpy as np

def func2(x):
    return 3*x*np.exp(x)
```

```python
f = lambda x : func2(x)
```

```python
ret3 = numerical_derivative(f, 2.0)
print(ret3)
```
