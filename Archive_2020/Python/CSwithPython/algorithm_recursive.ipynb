# Recursive 재귀 함수

## Factorial! : 재귀함수의 가장 기본 

```python
def fact(n):
    if n == 0:
        return 1
    else:
        return n * fact(n-1)
```

```python
fact(10)
```

- 함수는 호출할 때마다 함수가 새로이 실행된다.
- 이것은 각 함수 호출마다 인자 값을 포함한 모든 지역 변수가 새로이 만들어진다는 것을 의미한다.

## 문자열을 뒤집는 함수1

```python
def reverse(s):
    return reverse(s[1:]) + s[0]
```

```python
reverse('hello')
```

## 재귀함수가 순환함수가 안되려면!
- 재귀 호출이 필요하지 않은 기본 케이스가 필요하다. 

```python
def reverse(s):
    if s =="":
        return s
    else:
        return reverse(s[1:]) + s[0]
```

```python
reverse('hello')
```

```python
k = 'hello'
```

```python
reverse('si')
```

```python
reverse(k)
```
