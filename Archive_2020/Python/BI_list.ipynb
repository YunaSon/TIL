```python
vowels = ['a','e','i','o','u']
word = "Milliways"
found = []
for letter in word:
    if letter in vowels:
        if letter not in found:
            found.append(letter)
for vowel in found:
    print(vowel)
```

```python
paranoid_android = "Marvin, the Paranoid Android"
letters = list(paranoid_android)
```

```python
for char in letters[:6]:
    print('\t', char)
```

```python
for char in letters[-7:]:
    print('\t'*2, char)
```

```python
for char in letters[12:20]:
    print('\t'*3, char)
```
