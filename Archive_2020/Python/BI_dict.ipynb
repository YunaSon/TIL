## 딕셔너리
- 딕셔너리는 키/값 쌍을 저장합니다. (연관배열, 맵, 해시, 테이블)
- 딕셔너리 모델에 부합하는 데이터는 두개의 칼럼(열)과 여러 데이터 로우(행)을 갖는 것.
- 검색테이블: 여러분이 종이 사전에서 단어를 검색할 때 왼쪽에는 키 오른쪽에는 값을 검색하기 위해 사용됨

```python
person3 = {'Name': 'Ford prefect',
            'Gender': 'Male',
            'Occupation': 'Researcher',
            'Home Planet': 'Betelgeuse Seven'}
```

## 딕셔너리 빈도수

```python
found = {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 0}
```

```python
found['e'] = found['e'] + 1
```

```python
found
```

```python
found['a'] += 1
found
```

## 딕셔너리 반복

```python
### 1. 키
for i in found:
    print(i)
```

```python
for i in found.keys():
    print(i)
```

```python
### 2. value
for i in found:
    print(found[i])
```

```python
for i in found.values():
    print(i)
```

```python
##3. key와 value쌍 (item)
for i in found.items():
    print(i)
```

```python
##4. key, value쌍 unpacking
for i, j in found.items():
    print(i, j)
```

```python
fruits = {}
fruits
```

```python
fruits['apple'] = 10
fruits
```

```python
'apple' in fruits
```

### 인터프리터가 위 코드를 실행한 다음 메모리에 저장된 fruits를 어떻게 바라보는가...?

```python
if 'bananas' in fruits:
    fruits['bananas'] += 1
else:
    fruits['bananas'] = 1
```

```python
fruits
```

### dict의 key초기화: setdefault 메서드 // not in 연산자

```python
if 'pears' not in fruits:
    fruits['pears'] = 0
```

```python
fruits['pears'] += 1
fruits
```

```python
fruits.setdefault('cake', 0)
fruits['cake'] += 1
fruits
```
