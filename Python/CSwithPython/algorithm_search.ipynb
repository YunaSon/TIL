## 나의 풀이

```python
def search(x, nums):
    for i, j in enumerate(nums):
        if j == x:
            return i
    if x not in nums:
        return -1
```

## 선형 탐색

```python
def search(x, nums):
    for i, j in enumerate(nums):
        if j == x:
            return i
    return -1
```

```python
def search(x, nums):
    for i in range(len(nums)):
        if nums[i] == x:
            return i
    return -1
```

```python
search(3, nums)
```

## 이진 탐색

```python
def search(x, nums):
    low = 0
    high = len(nums) - 1
    while low <= high:
        mid = (low + high) //2
        item = nums[mid]
        if x == item:
            return mid
        elif x < item:
            high = mid - 1
        else:
            low = mid + 1
    return -1
```

```python
search(2, nums)
```
