### Class를 이용한 데이터 처리
- 객체는 어떤 사람이나 사물에 대한 정보를 모아놓은 용도로도 많이 쓰인다. 

- 레코드
    - 컴퓨터 과학에서 레코드(record, struct)는 기본적인 자료 구조이다. 데이터베이스나 스프레드시트의 레코드는 보통 로우(row)라고 부른다.

    - 레코드는 각기 다른 자료형에 속할 수 있는 필드의 모임이며, 보통 고정 숫자나 시퀀스로 이루어져 있다. 레코드의 필드들은 특히 객체 지향 프로그래밍에서 멤버(member)로도 부른다.

    - 예시. 회사의 인사 시스템: employee라는 객체가 있어 직원 이름, 주민등록번호, 주소, 급여, 소속부서

- 레코드 구현 예시 (C언어)
    ```
    struct date {
        int year;
        int month;
        int day;
    };
    ```

````python
class Student:
    def __init__(self, name, hours, qpoints):
        self.name = name
        self.hours = float(hours)
        self.qpoints = float(qpoints)
    
    def getName(self):
        return self.name

    def getHours(self):
        return self.hours
    
    def getQPoints(self):
        return self.qpoints

    def qpa(self):
        return self.qpoints/self.hours
````

### 평점이 가장 우수한 학생을 찾는 프로그램

- 사용자가 파일 이름을 입력 받는다. 
- 파일을 읽기 모드로 연다. 
- 첫 번째 학생을 최우수 학생으로 반복한다. 
- 각 학생 s에 대해 다음을 반복한다.
     ```
     if s.gpa() > best.gpa():
            s의 값을 새로운 best의 값으로 한다.
     ```
- best가 가리키는 학생에 대한 정보를 출력한다.
