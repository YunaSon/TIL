```python
# union
vowels = set('aeiou')
word = 'hello'
u = vowels.union(set(word))
u
```

```python
u_list = sorted(list(u))
u_list
```

```python
# difference
d = vowels.difference(set(word))
d
```

```python
# intersection
i = vowels.intersection(set(word))
i
```

```python
vowel = set('aeiou')
word = input("type the word: ")
result = vowel.intersection(set(word))
print(result)
```
