```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)

print(quicksort([3,6,8,10,1,2,1]))
```

```python
def solution(inputdata):
    xs = []
    ys = []
    result = []
    for i in range(len(inputdata)):
        xs.append(inputdata[i][0])
        ys.append(inputdata[i][1])
    if xs[0] != xs[1]:
        if xs[0] == xs[2]:
            result.append(xs[1])
        else:
            result.append(xs[0])
    else:
        result.append([2])
    if ys[0] != ys[1]:
        if ys[0] == ys[2]:
            result.append(ys[1])
        else:
            result.append(ys[0])
    else:
        result.append(ys[2])
    return result   
```
