## <font color='sky blue'> 선택정렬 (Selection Sort) </font>
- 선택 정렬은 배열된 자료의 값을 모두 비교한 후 가장 작은 값을 선택하여 배열의 첫 번째 자리의 값과 위치를 교환하면서 가장 작은 값부터 정렬시키는 방법이다. 

- 정렬해야 할 자료: 8, 4, 7, 3, 5, 2, 9
    - 1단계 : 8-4-7-3-5-2-9, 8과 2를 교환. 
    - 2단계 : 2-(4-7-3-5-8-9), 교환 없음.
    - 3단계 : 2-4-(7-3-5-8-9), 4와 3을 교환.
    - 4단계 : 2-3-4-(7-5-8-9), 교환 없음.
    - 5단계 : 2-3-4-7-(5-8-9), 7과 5를 교환.
    - 6단계 : 2-3-4-5-7-(8-9), 교환 없음.
    - 7단계 : 2-3-4-5-7-8-(9), 교환 없음. 

- 결과:2, 3, 4, 5, 7, 8, 9

```python
def SelectionSort(aList):
    for i in range(len(aList) - 1):
        min_position = i
        for k in range( i+1, len(aList)):
            if aList[k] < aList[min_position]:
                min_position = k
                
        if min_position !=i:
            tmp = aList[i]
            aList[i] = aList[min_position]
            aList[min_position] = tmp
```

## <font color='sky blue'> 삽입정렬 (Insertion Sort) </font>
- 삽입 정렬은 자료 배열의 모든 요소를 앞에서 부터 차래대로 정렬하는 방식으로 이미 정렬된 배열 부분의 내용과 자신의 값을 비교하며 자신의 위치를 찾아 삽입함으로써 정렬을 완성하는 알고리즘.
- 정렬해야 할 자료: 8, 4, 7, 3, 5, 2, 9
    - 1단계 : 4-8-7-3-5-2-9, 4와 8을 교환 
    - 2단계 : 4-7-8-3-5-2-9, 8과 7을 교환
    - 3단계 : 4-7-3-8-5-2-9, 3와 8을 교환.
    - 4단계 : 4-7-3-5-8-2-9, 5와 8을 교환.
    - 5단계 : 4-7-3-5-2-8-9, 2와 8을 교환.
    - 6단계 : 2-7-3-5-4-8-9, 2과 4를 교환.
    - 7단계 : 2-7-3-5-4-8-9, 7과 3을 교환.
    - 8단계 : 2-3-7-5-4-8-9, 7과 5를 교환.
    - 9단계 : 2-3-5-7-4-8-9, 7과 4를 교환.
    - 10단계 : 2-3-5-4-7-8-9, 5와 4를 교환.

- 결과:2, 3, 4, 5, 7, 8, 9

```python
def InsortionSort(aList):
    for i in range(1, len(aList)):
        current_val = aList[i]
        position = i
        while position > 0 and aList[position - 1] > current_val:
            aList[position] = aList[position - 1]
            position = position - 1
        if position != i:
            aList[position] = current_val
```

```python
aList = [8, 4, 7, 3, 5, 2, 9]
```

```python
InsortionSort(aList)
print(aList)
```
