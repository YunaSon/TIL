```python
import torch
import torch.nn 
import torch.optim
```

## <font color='navy'>Backpropagation

```python
X = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]]).to('cpu')
Y = torch.FloatTensor([[0], [1], [1], [1]]).to('cpu')
```

```python
# nn Layers (Neural Network Layers), weigh와 bias를 직접 선정. nn.Linear 2개를 사용. 
w1 = torch.Tensor(2, 2).to('cpu')
b1 = torch.Tensor(2).to('cpu')
w2 = torch.Tensor(2, 1).to('cpu')
b2 = torch.Tensor(1).to('cpu')
```

```python
linear1 = torch.nn.Linear(2, 2, bias = True) # MLP
linear2 = torch.nn.Linear(2, 1, bias = True) # MLP
sigmoid = torch.nn.Sigmoid()
model = torch.nn.Sequential(linear1, sigmoid, linear2, sigmoid).to('cpu')

criterion = torch.nn.BCELoss().to('cpu')
optimizer = torch.optim.SGD(model.parameters(), lr=1)

for step in range(10001):
    optimizer.zero_grad()
    hypothesis = model(X)
    
    cost = criterion(hypothesis, Y)
    cost.backward()
    optimizer.step()
    if step % 100 == 0:
        print(step, cost.item())
```

```python
def sigmoid(x):
    return 1.0 / (1.0 + torch.exp(-x))
```

```python
def sigmoid_prime(x):
    return sigmoid(x) * (1-sigmoid(x))
```

```python
for step in range(10001):
    #forward = layer의 통과!!!
    l1 = torch.add(torch.matmul(X, w1), b1)
    a1 = sigmoid(l1)
    l2 = torch.add(torch.matmul(a1, w2), b2)
    Y_pred = sigmoid(l2)
    
    cost = -torch.mean(Y * torch.log(Y_pred) + (1 - Y) * torch.log(1 - Y_pred)) # binary cross entropy loss
```

```python
# forward!
# Backpropagation (Chain Rule)
for step in range(10001):
    # Loss derivative 
    d_Y_pred = (Y_pred - Y) / (Y_pred * (1.0 - Y_pred) + 1e-7) #binary cross entropy미분
    
    # Layer 2
    d_l2 = d_Y_pred * sigmoid_prime(12) # 시그모이드 함수 미분
    d_b2 = d_l2 # bias 미분
    d_w2 = torch.matmul(torch.transpose(a1, 0, 1), d_b2) # weigh 미분
    
    # Layer 1
    d_a1 = torch.matmul(d_b2, torch.transpose(w2, 0, 1))
    d_l1 = d_a1 * sigmoid_prime(l1)
    d_b1 = d_l1
    d_w1 = torch.matmul(torch.transpose(X, 0, 1), d_b1)
    
    #====================back propagation 끝
    #Weight Update
    w1 = w1 - learning_rate * d_w1
    b1 = b1 - learning_rate * torch.mean(d_b1, 0)
    w2 = w2 - learning_rate * d_w2
    b2 = b2 - learning_rate * torch.mean(d_b2, 0)
    
    if step % 100 == 0:
        print(step, cost.item())
```

### 4MLP,wide-deep

```python
linear1 = torch.nn.Linear(2, 10, bias = True)
linear2 = torch.nn.Linear(10, 10, bias = True)
linear3 = torch.nn.Linear(10, 10, bias = True)
linear4 = torch.nn.Linear(10, 1, bias = True)
sigmoid = torch.nn.Sigmoid()
```

```python
model = torch.nn.Sequential(linear1, sigmoid, linear2, sigmoid, linear3, sigmoid, linear4, sigmoid).to('cpu')

criterion = torch.nn.BCELoss().to('cpu')
optimizer = torch.optim.SGD(model.parameters(), lr=1)

for step in range(100):
    optimizer.zero_grad()
    hypothesis = model(X)
    
    cost = criterion(hypothesis, Y)
    cost.backward()
    optimizer.step()
    print(step, cost.item())
```
