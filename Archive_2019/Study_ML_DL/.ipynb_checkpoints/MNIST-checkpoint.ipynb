```python
import torchvision.datasets as dsets
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
import torch
import torch.nn
```

```python
mnist_train = dsets.MNIST(root="MNIST_data/", train=True, transform=transforms.ToTensor(), download=True)
mnist_test =dsets.MNIST(root="MNIST_data/", train=False, transform=transforms.ToTensor(), download=True)
```

```python
data_loader = DataLoader(mnist_train, batch_size=100, shuffle=True, drop_last=True)
```

```python
for X, Y in data_loader:
    X = X.view(-1, 28 * 28) # batchsize, W*H
```

```python
X.shape
```

```python
len(data_loader)
```

## Softmax

```python
linear = torch.nn.Linear(784, 10, bias = True).to('cpu')

#parameter
training_epochs = 15
batch_size = 100

#define cost/Loss & Optimizer
criterion = torch.nn.CrossEntropyLoss().to('cpu')
optimizer = torch.optim.SGD(linear.parameters(), lr=0.1)
```

```python
for epoch in range(training_epochs):
    # X: мқҙлҜём§Җ, Y: Label
    # reshape input image into (batch_size by 784)
    # Label is not one-hot encoded
    avg_cost = 0
    total_batch = len(data_loader)
    for X, Y in data_loader: #batchsizeл§ҢнҒј data_loaderлЎң н•ҷмҠө
        X = X.view(-1, 28*28).to('cpu')
        optimizer.zero_grad()
        hypothesis = linear(X)
        cost = criterion(hypothesis, Y)
        cost.backward() #backwardлҘј мқҙмҡ©н•ҙм„ң gradientлҘј кі„мӮ°
        optimizer.step() #optimizerм—җ update
        avg_cost += cost / total_batch
    
    print("Epoch: ", "%04d" % (epoch+1), "cost=", "{:.9f}".format(avg_cost))
```

## Test

```python
with torch.no_grad():
    X_test = mnist_test.test_data.view(-1, 28*28).float().to('cpu')
    Y_test = mnist_test.test_labels.to('cpu')
    
    prediction = linear(X_test)
    correct_prediction = torch.argmax(prediction, 1) == Y_test
    accuracy = correct_prediction.float().mean()
    print("Accuracy: ", accuracy.item())
```

## Epoch, BatchSize, Iteration
#### In the neural network terminology:
    - one epoch: one forward pass and one backward pass of all the training examples
    - batch size: the number of training examples in one forward/backward pass. The higher the batch size, the more memory space you'll need.
    - number of iterations: л°°м№ҳлҘј лӘҮлІҲ н•ҷмҠөм—җ мӮ¬мҡ© н•ҳмҳҖлӮҳ? number of passes, each pass using(batch size) number of examples. 
    
### ex. 
1000к°ңмқҳ datasetмқ„ batch size 500мңјлЎң лӮҳлҲ„кі  (batchмқҳ к°ңмҲҳлҠ” 2) 2нҡҢ iterationsн•ҳл©ҙ 1к°ңмқҳ epochмқҙ лҸҢм•ҳлӢӨ.
