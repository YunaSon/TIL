```python
import torch
import torch.nn as nn
import torchvision.datasets as dsets
from torch.utils.data import DataLoader
import torchvision.transforms as transforms
```

```python
linear1 = nn.Linear(784, 512, bias = True).to('cpu')
linear2 = nn.Linear(512, 512, bias = True).to('cpu')
linear3 = nn.Linear(512, 512, bias = True).to('cpu')
linear4 = nn.Linear(512, 512, bias = True).to('cpu')
linear5 = nn.Linear(512, 10, bias = True).to('cpu')
relu = nn.ReLU()
dropout = nn.Dropout(p=0.5)
```

```python
model = torch.nn.Sequential(linear1, relu, dropout, 
                            linear2, relu, dropout, 
                            linear3, relu, dropout, 
                            linear4, relu, dropout, 
                            linear5).to('cpu')
```

```python
model.train() # set the model to train mode( Dropout=True)

with torch.no_grad():
    model.eval() # set the model to evaluation mode (Dropout=False)
```
