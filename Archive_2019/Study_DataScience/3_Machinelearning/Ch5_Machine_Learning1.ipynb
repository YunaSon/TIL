## <font color='rainbow'> What is Machine Learning? </font>

- 데이터를 이해하는 데 도움이 되는 수학적 모델을 구축하는 것. 
- 관측된 데이터에 적용할 수 있는 조정 가능한 매개 변수(parameter)를 구축한 모델에 제공하면 "학습"이 시작된다.

**"기계 학습은 명시적으로 프로그래밍되지 않고 컴퓨터가 작동하도록하는 과학입니다."**

다른 곳의 정의
- 머신 러닝 기술은 명시적으로 프로그래밍하는 것이 아니라 데이터로부터의 학습을 통해 작업을 수행하는 방법을 컴퓨터에 가르치는 것
- 머신 러닝은 사용하는 데이터를 기반으로 학습하거나 성능을 개선하는 시스템을 구축하는 데 초점이 맞춰져 있는 인공 지능(AI)의 하위집합입니다.

## <font color='rainbow'> Categories of Machine Learning </font>

### <font color='darkgreen'>  Supervised Learning </font>

지도 학습은 측정된 데이터의 특징과 데이터와 관련된 일부 레이블(정답?) 사이의 관계를 모델링함. 
<br>모델이 결정되면 새로운, 알 수 없는 데이터의 레이블을 적용하는 데 사용할 수 있다. 예) Classification and regression </br>

### <font color='darkgreen'>  Unsupervised Learning </font>

비지도 학습은 레이블을 참조하지 않고 데이터 집합의 feature를 모델링하는 것과 관련이 있으며, 
<br>종종 "데이터 집합이 스스로 말하게하는 것"으로 설명됨</br> 
<br> 예) Clustering and Dimensionality reduction </br>

### <font color='darkgreen'>  Semi-supervised Learning </font>

반지도 학습 방법은 지도 학습과 비지도 학습 사이에 있음. 반지도 학습 방법은 불완전한 레이블만 사용할 수 있는 경우에 유용함.

### <font color='darkgreen'>  Supervised learning example: Simple linear regression </font>

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
plt.scatter(x, y);
```

### <font color='black'>  1. Choose a class of model </font>

Scikit-Learn에서 모든 모델 클래스는 Python 클래스로 표시된다. 
<br>예를 들어 간단한 선형 회귀 모델을 계산하려면 선형 회귀 클래스를 가져올 수 있음</br>

```python
from sklearn.linear_model import LinearRegression
```

### <font color='black'> 2. Choose model hyperparameters </font>

중요한 점은 **모델 클래스가 모델 인스턴스와 동일하지 않다**는 것

모델 클래스를 결정한 후에도 몇 가지 옵션이 있음
<br>작업중인 모델 클래스에 따라 다음과 같은 하나 이상의 질문에 대답해야 할 수도 있음</br>

- 오프셋에 적합합니까 (예 : y 절편)?
- 모델을 정규화 하시겠습니까?
- 모델 유연성을 추가하기 위해 기능을 사전 처리하고 싶습니까?
- 모델에서 어느 정도의 정규화를 사용 하시겠습니까?
- 몇 개의 모델 구성 요소를 사용 하시겠습니까?

하이퍼파라미터(hyperparameter)란? 
<br>모델 학습 전에 미리 지정해야 하는 학습 알고리즘이 자체적으로 가지고 있는 파라미터</br>

**Scikit-Learn에서는 모델 인스턴스화시 값을 전달하여 hyperparameter가 선택됨**

선형 회귀 예제의 경우 LinearRegression 클래스를 인스턴스화하고 fit_intercept 하이퍼 파라미터를 사용하여 절편을 맞추도록 지정할 수 있음

<font color='red'>모델의 option들.. hyperparameter...</font>

```python
model = LinearRegression(fit_intercept=True)
model
```

모델이 인스턴스화 될 때 유일한 조치는 이러한 하이퍼 파라미터 값을 저장하는 것
<br>Scikit-Learn API는 모델 선택과 모델의 데이터 적용을 구분함</br>

### <font color='black'> 3. Arrange data into a features matrix and target vector </font>

데이터 x를 크기 [n_samples, n_features]의 행렬로 만들어야 함 

```python
X = x[:, np.newaxis]
X.shape
```

### <font color='black'> 4. Fit the model to your data </font>

- 모델을 데이터에 적용 
- 모델의 fit () 메소드로 수행

<font color='red'>모델을 학습한다. = fitting한다. </font>

```python
model.fit(X, y)
```

- fit () 명령을 사용하면 여러 모델 종속 내부 계산이 수행되며, 계산 결과는 사용자가 탐색 할 수있는 모델 별 속성에 저장됨 
- Scikit-Learn에서 fit () 프로세스 중에 학습 된 모든 model parameter는 밑줄이 있음

```python
model.coef_
```

```python
model.intercept_
```

### <font color='black'> 5. Predict labels for unknown data </font>

<br> predict () 메소드를 사용하여 unknown data에 대한 평가를 수행</br> 

```python
xfit = np.linspace(-1, 11)
```

### <font color='darkgreen'> Supervised learning example: Iris classification </font>

Question) 아이리스 데이터의 일부에 대해 훈련된 모델이 있다면, 남은 라벨을 얼마나 잘 예측할 수 있을까?

- Gaussian Naive Bayes을 사용 
- 이 모델은 각 클래스가 axis-aligned Gaussian 분포에서 추출되었다고 가정
- 매우 빠르며 선택할 수 있는 hyperparameter가 없음

```python
from sklearn.model_selection import train_test_split  # sklearn.cross_validation is deprecated.
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris,
                                                random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB # 1. choose model class
model = GaussianNB()                       # 2. instantiate model
model.fit(Xtrain, ytrain)                  # 3. fit model to data
y_model = model.predict(Xtest)             # 4. predict on new data
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

<br>각 샘플에 대해 4 개의 feature가 기록됨</br>

차원 축소 작업은 데이터의 필수 기능을 유지하는 적절한 저 차원 표현이 있는지 묻는 것
<br>차원 축소는 데이터 시각화를 위한 보조 수단으로도 사용됨</br> 
<br>4 차원 이상에서 보다 2 차원으로 데이터를 작성하는 것이 훨씬 쉽다!</br>

principal component analysis(PCA)방법을 이용 
<br>데이터의 2 차원 표현을 반환하도록 요청</br>

```python
from sklearn.decomposition import PCA  # 1. Choose the model class
model = PCA(n_components=2)            # 2. Instantiate the model with hyperparameters
model.fit(X_iris)                      # 3. Fit to data. Notice y is not specified!
X_2D = model.transform(X_iris)         # 4. Transform the data to two dimensions
```

이를 수행하는 빠른 방법은 결과를 원본 Iris DataFrame에 삽입하고 Seaborn의 lmplot을 사용하여 결과를 표시

```python
iris['PCA1'] = X_2D[:, 0]
iris['PCA2'] = X_2D[:, 1]
sns.lmplot("PCA1", "PCA2", hue='species', data=iris, fit_reg=False);
```

PCA 알고리즘에 species lable에 대한 지식이 없더라도 2 차원 표현에서 종이 상당히 잘 분리되어 있음을 알 수 있음!

### <font color='darkgreen'> Unsupervised learning: Iris clustering </font>

Iris 데이터에 클러스터링을 적용
<br>클러스터링 알고리즘은 label을 참조하지 않고 별개의 데이터 그룹을 찾고자 함.</br> 
<br>여기서는 In Depth : Gaussian Mixture Models에 자세히 설명되어 있는 Gaussian 혼합 모델 (GMM)이라는 방법을 사용</br> 
<br>GMM은 데이터를 Gaussian blob 모음으로 모델링하려고 함</br>

```python
from sklearn.mixture import GaussianMixture      # 1. Choose the model class sklearn.mixture.GaussianMixture
model = GaussianMixture(n_components=3,
            covariance_type='full')  # 2. Instantiate the model with hyperparameters
model.fit(X_iris)                    # 3. Fit to data. Notice y is not specified!
y_gmm = model.predict(X_iris)        # 4. Determine cluster labels
```

```python
iris['cluster'] = y_gmm
sns.lmplot("PCA1", "PCA2", data=iris, hue='species',
           col='cluster', fit_reg=False);
```

## <font color='rainbow'> Thinking about Model Validation </font> 

모델 검증에 대한 순진한 접근 방식과 왜 실패하는지를 보며,
<br>보다 견고한 모델 평가를 위해 holdout set 및 cross-validation을 이용해 본다</br>

### <font color='darkgreen'> Model validation the wrong way </font> 

```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
```

model과 hyperparameter를 선택
<br>k-neighbors classifier를 사용. option - n_neighbors=1</br>
<br>model의 가정 - "unknown point의 label은 가장 가까운 training point의 label과 동일하다."</br>

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=1)
```

```python
model.fit(X, y)
y_model = model.predict(X)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y, y_model)
```

**100% 일치라니 말이되나?**
<br>위의 방법의 문제는?</br>
<br>>>동일한 데이터에서 모델을 훈련하고 평가했음</br>
<br>nearest neighbor model은 훈련 데이터를 간단히 저장하는 instance-based estimator이며, 새로운 데이터를 저장된 점과 비교하여 label을 예측함 
<br>조작된 경우를 제외하고는 매번 100% 정확도를 얻는다!</br>

### <font color='darkgreen'>  Model validation the right way: Holdout sets </font>

<br>model training에서 데이터의 일부 하위 집합을 보류한 다음 holdout set를 사용하여 모델 성능을 확인함</br>
<br>이 분할은 Scikit-Learn의 train_test_split 유틸리티를 사용하여 수행 할 수 있다</br>

```python
from sklearn.model_selection import train_test_split
# split the data with 50% in each set
X1, X2, y1, y2 = train_test_split(X, y, random_state=0,
                                  train_size=0.5)

# fit the model on one set of data
model.fit(X1, y1)

# evaluate the model on the second set of data
y2_model = model.predict(X2)
accuracy_score(y2, y2_model)
```

### <font color='darkgreen'>  Model validation via cross-validation </font>

모델 검증에 holdout set을 사용할 때의 단점
<br>>> 모델 훈련에서 데이터의 일부를 사용하지 못함</br> 
<br>위의 경우 데이터 세트의 절반이 loss! </br>
<br>초기 훈련 데이터 세트가 작은 경우 문제가 됨</br>

이를 해결하는 한 가지 방법은 cross-validation를 사용하는 것
<br>데이터의 각 subset을 training set 및 validation set으로 사용</br>

여기서는 데이터의 각 절반을 holdout set로 번갈아 사용하여 두 가지 유효성 검사를 수행

```python
y2_model = model.fit(X1, y1).predict(X2)
y1_model = model.fit(X2, y2).predict(X1)
accuracy_score(y1, y1_model), accuracy_score(y2, y2_model)
```

이 아이디어를 확장하여 더 많은 시도와 더 많은 데이터를 사용할 수 있음
<br>예를 들어, 5 배 교차 검증을 시도</br>

<font color='red'>bootstraping</font>

<img src="./figures/ml14.png" alt="drawing" width="600"/>

데이터를 5 개의 그룹으로 나누고 각 그룹을 차례로 사용하여 데이터의 다른 4/5에 맞는 모델을 평가함

```python
from sklearn.model_selection import cross_val_score
cross_val_score(model, X, y, cv=5)
```

Scikit-Learn은 특정 상황에서 여러 가지 유용한 cross-validation을 구현할 수 있음 
<br>>> cross_validation module의 iterator를 통해 구현됨</br>

**Leave-one-out cross-validation**
<br>예를 들어, fold number가 데이터 포인트 수와 동일한 극단적인 경우로 수행할 수 있음 </br>
<br>즉, 각 시행에서 하나씩 만 모든 포인트에 대해 training을 진행</br> 
<br>이 유형을 leave-one-out 교차 유효성 검사라고 함</br>
<br>총 sample중 1개만을 test set으로 사용하여, sample 수 만큼 검증을 진행</br>

```python
from sklearn.model_selection import LeaveOneOut
scores = cross_val_score(model, X, y, cv=LeaveOneOut())
scores
```

150 개의 표본이 있으므로, 1 회 교차 검증에서 150 번의 시험에 대한 점수가 산출되며 
<br>점수는 성공 (1.0) 또는 실패 (0.0) 예측을 나타냅니다. 이를 평균하면 오류율이 추정됨</br>

```python
scores.mean()
```

### <font color='darkgreen'>  The Bias-variance trade-off </font>

기본적으로 "최적의 모델"을 찾는 문제는 편향(bias)과 분산(variance) 사이의 상충 관계에서 적절한 지점을 찾는 것이다

1. 편향(Bias)
<br>일반적인 용어로 추정량의 편향은 기대값(E)과 추정되어야할 진짜 값 사이의 차이임</br>

2. 분산(Variance)
<br>추정량과 기대값의 단순한 통계적 분산을 말함</br>
<br>모델 구축 프로세스가 데이터의 변화에 민감할수록 분산은 더 커짐</br>

<font color='red'>
    <ul>-Bias:몬테카를로, 학습할때마다 모델이 업데이트, 학습을 안한거...</ul>
    <ul>-Variance: 너무 학습한거...</ul></font>

<img src="./figures/ml15.png" alt="drawing" width="800"/>

왼쪽의 모형은 데이터를 통해 직선 일치를 찾고자 합니다. 
<br>데이터가 본질적으로 직선보다 복잡하기 때문에 직선 모델은 이 데이터 세트를 잘 설명 할 수 없습니다.</br> 
<br>이러한 모델은 데이터에 적합하지 않습니다.</br> 
<br>즉, 데이터의 모든 기능을 적절하게 설명할 수 있는 충분한 모델 유연성이 없습니다.</br> 
<br>이것을 말하는 또 다른 방법은 **"모델의 편향(bias)이 높다"**는 것입니다.</br>

오른 쪽의 모형은 데이터를 통해 고차 다항식을 맞추려고 합니다. 
<br>여기서 모델 적합은 데이터의 미세한 특징을 거의 완벽하게 설명하기에 충분한 유연성을 갖지만, </br>
<Br>훈련 데이터를 매우 정확하게 설명하더라도 정확한 형식은 본질적인 것이 아니라,</br>
<br>**데이터의 특정 노이즈 특성을 더 반영하는 것으로 보입니다.** </br>

<br>해당 데이터를 생성한 프로세스의 속성, 즉 model은 데이터를 과적합(overfit) 한다고 합니다. </br>
<br>즉, 모델 유연성이 너무 높아서 모델이 임의의 오류와 기본 데이터 분포를 설명하게 됩니다. </br>
<br>이것은 **"모델의 분산(variance)이 높다"** 는 것입니다.</br>

<font color ='red'>
    <ul>- 회귀편차 : </ul>  
    <ul>- 오차(잔차) : </ul>
    <ul>- 총편차 : </ul>

SST(분산) = SSR + SSE
R^2 = SSR/SST, 결정계수: 예측력이 높다.

</font>

<br>다음 다이어그램에서 빨간색 / 밝은 점은 학습 세트에서 생략된 데이터를 나타낸다</br>

점수는 $R^2$ 점수 또는 결정 계수(coefficient of determination)로, model의 성능이 target value의 단순 평균과 얼마나 비슷한지를 측정함
<br>$R^2$ = 1은 완벽하게 일치함을 나타내고 </br>

- 고 분산 모델의 경우 검증 세트의 모델 성능이 훈련 세트의 성능보다 훨씬 나쁘다

여기에 표시된 다이어그램을 validation curve라고하며 다음과 같은 특징이 있다

- training score는 validation score보다 일반적으로 높다
<br>모델은 보지 않은 데이터보다 보았던 데이터에 더 적합하다</br>
- 모델 복잡성이 매우 낮은 경우 (고 편향 모델) 학습 데이터가 적합하지 않으므로, 
<br>모델은 train data와 test data 모두에 대한 예측이 좋지 않다</br>
- 매우 복잡한 모델 (고 분산 모델)의 경우 훈련 데이터가 overfit되므로, 
<br>모델이 train data를 매우 잘 예측하지만 test data에 대한 예측은 실패한다</br>
- 일부 중간 값의 경우 유효성 검사 곡선이 최대치가 됨 
<br>이 수준의 complexity는 바이어스와 분산 사이의 적절한 균형을 이룬다</br>

### <font color='darkgreen'> Validation curves in Scikit-Learn </font>

cross-validation를 사용하여 모델 클래스의 유효성 검사 곡선을 계산해보자!
<br>>> 다항식 회귀 모형(polynomial regression model)을 사용 </br>
<br>이것은 다항식의 정도가 조정 가능한 매개 변수(tunable parameter)인 일반화된 선형 모형이다. </br>

- 1차 함수 : y=ax+b (직선에 적합)
- 3차 함수 : y=ax3+bx2+cx+d (3차 곡선에 적합) 
-<font color='red'>Feature는 1개. 연관성이 3차.</font>

Scikit-Learn에서는 다항식 전처리기와 결합된 간단한 선형회귀로 구현 가능함
<br>파이프 라인을 사용하여 이러한 작업을 함께 묶는다</br>

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

def PolynomialRegression(degree=2, **kwargs):
    return make_pipeline(PolynomialFeatures(degree),
                         LinearRegression(**kwargs))
```

```python
import numpy as np

def make_data(N, err=1.0, rseed=1):
    # randomly sample the data
    rng = np.random.RandomState(rseed)
    X = rng.rand(N, 1) ** 2
    y = 10 - 1. / (X.ravel() + 0.1)
    if err > 0:
        y += err * rng.randn(N)
    return X, y

X, y = make_data(40)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()  # plot formatting

X_test = np.linspace(-0.1, 1.1, 500)[:, None]

plt.scatter(X.ravel(), y, color='black')
axis = plt.axis()
for degree in [1, 3, 5]:
    y_test = PolynomialRegression(degree).fit(X, y).predict(X_test)
    plt.plot(X_test.ravel(), y_test, label='degree={0}'.format(degree))
plt.xlim(-0.1, 1.0)
plt.ylim(-2, 12)
plt.legend(loc='best');
```

이 경우 모델 복잡성을 제어하는 노브는 다항식의 차수이며, 이는 음수가 아닌 정수일 수 있다.
<br>**(Question)**</br>
<br>어느 정도의 다항식이 편향 (under-fitting)과 분산 (over-fitting)간에 적절한 균형을 이룰 수 있나? </br>

- 특정 데이터 및 모델에 대한 유효성 검사 곡선을 시각화하여 진행할 수 있다
- Scikit-Learn에서 제공하는 validation_curve의 루틴을 사용하여 진행가능
- 모델, 데이터, 매개 변수 이름 및 탐색할 범위가 주어지면,</br>
- 이 함수는 범위에서 학습 점수와 유효성 검사 점수를 자동으로 계산</br>
-<font color='red'>degree를 보고 over fitting여부...생각할수 있다.</font>

```python
from sklearn.model_selection import validation_curve
degree = np.arange(0, 21)
train_score, val_score = validation_curve(PolynomialRegression(), X, y,
                                          'polynomialfeatures__degree', degree, cv=7)

plt.plot(degree, np.median(train_score, 1), color='blue', label='training score')
plt.plot(degree, np.median(val_score, 1), color='red', label='validation score')
plt.legend(loc='best')
plt.ylim(0, 1)
plt.xlabel('degree')
plt.ylabel('score');
```

- trainging score는 validation score보다 어디서든 높다 
- trainging score는 모델 복잡성이 증가하면 조금씩 향상된다 </br>
- model이 over-fitting 됨에 따라 validation score는 최대치에 도달한 이후 떨어지게 된다</br>

검증 곡선(validation curve)으로부터, 3차 다항식에서 바이어스와 분산 사이의 최적 트레이드 오프가 일어난 것을 알수 있음
<br>다음과 같이 원본 데이터에 대해 적합치를 계산하고 표시 할 수 있다</br>

```python
plt.scatter(X.ravel(), y)
lim = plt.axis()
y_test = PolynomialRegression(3).fit(X, y).predict(X_test)
plt.plot(X_test.ravel(), y_test);
plt.axis(lim);
```

최적 모델을 찾는데 실제로 훈련 점수를 계산할 필요는 없지만,
<br>훈련 점수와 유효성 검사 점수 사이의 관계를 조사하면 모델 성능에 대한 유용한 통찰력을 얻을 수 있습니다.</br>

따라서 검증 곡선의 동작에는 **모델 복잡성과 훈련 포인트 수**라는 두 가지 중요한 입력이 있다 
<br>모델에 맞게 점점 더 큰 데이터 하위 집합을 사용하여 수행할 수 있는 </br>
<br>훈련 포인트 수의 함수로 모델의 동작을 탐색하는 것이 때로는 유용함 </br>
<br>training set의 크기에 대한 traing / validation score plot을 학습 곡선(learning curve)라고 함</br>
<br> ==> training set의 수를 늘려가며 plot을 확인

 Learning curve에서 기대할 수 있는 점

- 주어진 복잡성의 모델은 작은 데이터 세트에는 over-fit하게 됨
<br>즉, training score는 상대적으로 높고 validation score는 상대적으로 낮다</br>
- 주어진 복잡성의 모델은 큰 데이터 세트에 적합하지 않습니다. 
<br>즉, training score는 줄어들지만 validation score는 증가함</br>
- 우연한 경우를 제외하고 모델은 training set보다 validation set의 점수가 낮다

Learning curve의 주목할만한 특징은 훈련 샘플 수가 증가함에 따라 특정 점수로 수렴하는 것
<br>특히, 특정 모델이 수렴할만큼 충분한 포인트가 있으면 더 많은 교육 데이터를 추가해도 도움이되지 않는다! 
<br>이 경우 모델 성능을 향상시키는 유일한 방법은 **다른 (종종 더 복잡한) 모델을 사용하는 것**이다</br>

### <font color='darkgreen'> Learning curves in Scikit-Learn </font>

2차 다항식 모델과 9차 다항식을 사용하여 원래 data set에 대한 learning curve를 계산해보자

```python
from sklearn.model_selection import learning_curve

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)

for i, degree in enumerate([2, 9]):
    N, train_lc, val_lc = learning_curve(PolynomialRegression(degree),
                                         X, y, cv=7,
                                         train_sizes=np.linspace(0.3, 1, 25))

    ax[i].plot(N, np.mean(train_lc, 1), color='blue', label='training score')
    ax[i].plot(N, np.mean(val_lc, 1), color='red', label='validation score')
    ax[i].hlines(np.mean([train_lc[-1], val_lc[-1]]), N[0], N[-1],
                 color='gray', linestyle='dashed')

    ax[i].set_ylim(0, 1)
    ax[i].set_xlim(N[0], N[-1])
    ax[i].set_xlabel('training size')
    ax[i].set_ylabel('score')
    ax[i].set_title('degree = {0}'.format(degree), size=14)
    ax[i].legend(loc='best')
```

이것은 모델이 훈련 데이터 증가에 어떻게 반응하는지 시각적으로 보여 주므로 유용하다
<br>특히 learning curve가 이미 수렴 된 경우 (즉, 훈련 및 유효성 검사 곡선이 이미 서로 가까이있는 경우),</br>
<br>더 많은 교육 데이터를 추가해도 적합도가 크게 향상되지 않음! </br>

수렴 점수를 높이는 유일한 방법은 다른 (보통 더 복잡한) 모델을 사용하는 것이다
<br>훨씬 더 복잡한 모델로 이동하면 수렴 점수 (점선으로 표시)가 증가하지만,</br>
<br>더 높은 모델 분산 (훈련 점수와 검증 점수의 차이로 표시)을 잃게된다</br> 
<br>더 많은 데이터 포인트를 추가한다면 더 복잡한 모델에 대한 학습 곡선이 결국 수렴 될 것임</br>

## <font color='rainbow'> Derived Features </font> 

일부 input feature에서 수학적으로 파생된 feature임
<br>입력 데이터에서 polynomial feature를 구성 할 때 Hyperparameters and Model Validation에서 이에 대한 예를 보았음</br> 
<br>모델을 변경하는 것이 아니라 입력을 변환하여 simple regression를 polynomial regression로 변환 할 수 있다 </br>
<br>이것을 basis function regression라고도 함</br>

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y = np.array([4, 2, 1, 3, 7])
plt.scatter(x, y);
```

여전히 LinearRegression을 사용하여 데이터에 선을 맞추고 최적의 결과를 얻을 수 있다

```python
from sklearn.linear_model import LinearRegression
X = x[:, np.newaxis]
model = LinearRegression().fit(X, y)
yfit = model.predict(X)
plt.scatter(x, y)
plt.plot(x, yfit);
```

이에 대한 한 가지 접근 방식은 데이터를 변환하고 feature column을 추가하여 모델의 유연성을 높이는 것
<br>예를 들어, 다음과 같이 데이터에 다항식 기능을 추가 할 수 있음</br>

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=3, include_bias=False)
X2 = poly.fit_transform(X)
print(X2)
```

Drived feature matrix에는 x를 나타내는 하나의 열과 $x^2$를 나타내는 두 번째 열과 $x^3$을 나타내는 세 번째 열이 있음 
<br>이 확장된 입력에서 linear regression를 계산하면 데이터에 훨씬 더 적합함</br>

```python
model = LinearRegression().fit(X2, y)
yfit = model.predict(X2)
plt.scatter(x, y)
plt.plot(x, yfit);
```

모델을 변경하는 것이 아니라 입력을 변환하여 모델을 개선한다는,
<br>이러한 아이디어는 보다 강력한 머신 러닝 방법의 기본</br>

## <font color='rainbow'> Imputation of Missing Data </font> 

Feature engineering의 또 다른 일반적인 요구는 누락된 데이터를 처리하는 것

```python
from numpy import nan
X = np.array([[ nan, 0,   3  ],
              [ 3,   7,   9  ],
              [ 3,   5,   2  ],
              [ 4,   nan, 6  ],
              [ 8,   8,   1  ]])
y = np.array([14, 16, -1,  8, -5])
```

이러한 데이터에 일반적인 기계 학습 모델을 적용할 때는,
<br>먼저 누락된 데이터를 적절한 채우기 값으로 대체해야 함 </br>
<br>이것은 결측값의 대치라고 알려져 있으며 전략은 단순한 것 </br>
<br>(예 : 결측값을 열의 평균으로 대체)에서 </br>
<br>정교함 (예 : 행렬 완성 또는 이러한 데이터를 처리하기위한 강력한 모델 사용)까지 다양함</br>

정교한 접근 방식은 응용 프로그램마다 매우 다른 경향이 있음
<br>평균, 중간 또는 가장 빈번한 값을 사용하는 기본적인 접근 방식의 경우,</br>
<br>Scikit-Learn은 Imputer 클래스를 제공</br>

```python
#from sklearn.preprocessing import Imputer
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean') # function method is changed
X2 = imp.fit_transform(X)
X2
```

```python
model = LinearRegression().fit(X2, y)
model.predict(X2)
```

## <font color='rainbow'> Feature Pipelines </font> 

앞의 예제를 사용할 때 특히 여러 단계를 함께 묶으려는 경우,
<br>preprecessing pipeline으로 일괄처리 가능</br>

- 평균을 사용하여 결측값 대치
- feature를 2차로 변환
- Fit a linear regression

```python
from sklearn.pipeline import make_pipeline

model = make_pipeline(SimpleImputer(missing_values=np.nan, strategy='mean'), # function method is changed
                      PolynomialFeatures(degree=2),
                      LinearRegression())
```

```python
model.fit(X, y)  # X with missing values, from above
print(y)
print(model.predict(X))
```
