## <font color='orange'>머신러닝이란?
- 데이터를 이해하는 데 도움이 되는 수학적 모델을 구축하는 것. 
- 관측된 데이터에 적용할 수 있는 조정 가능한 매개 변수(parameter)를 구축한 모델에 제공하면 "학습"이 시작된다.
- 머신 러닝은 종종 AI의 한 분야로 분류되는데 그것보다는 data에서 정보를 얻을 수 있도록 model을 설계하는것
    
**"기계 학습은 명시적으로 프로그래밍되지 않고 컴퓨터가 작동하도록하는 과학"** </br>

**"Machine Learning as a means of building models of data"**

다른 곳의 정의
- 머신 러닝 기술은 명시적으로 프로그래밍하는 것이 아니라 데이터로부터의 학습을 통해 작업을 수행하는 방법을 컴퓨터에 가르치는 것
- 머신 러닝은 사용하는 데이터를 기반으로 학습하거나 성능을 개선하는 시스템을 구축하는 데 초점이 맞춰져 있는 인공 지능(AI)의 하위집합입니다.

## <font color='orange'> Categories of Machine Learning </font>

### <font color='navy'> Supervised Learning </font>
- 지도 학습은 측정된 데이터의 특징과 데이터와 관련된 일부 레이블(정답) 사이의 관계를 모델링함. 
- 모델이 결정되면 새로운, 알 수 없는 데이터의 레이블을 적용하는 데 사용할 수 있다. 
- Classification, Regression 

### <font color='navy'> Unsupervised Learning </font>
- 비지도 학습은 레이블을 참조하지 않고 데이터 집합의 feature를 모델링하는 것과 관련이 있으며, 
- 종종 "데이터 집합이 스스로 말하게하는 것"으로 설명됨</br> 
- Clustering, Dimensionality reduction 

### <font color='navy'> Semi-supervised Learning </font>
- 반지도 학습 방법은 지도 학습과 비지도 학습 사이에 있음. 반지도 학습 방법은 불완전한 레이블만 사용할 수 있는 경우에 유용함.

## <font color='orange'>  요약 </font>

#### Supervied learning : 레이블이 있는 학습 데이터를 기반으로 레이블을 예측할 수있는 모델
- Classification : 레이블을 둘 이상의 개별 범주로 예측하는 모델
- Regression : 연속 레이블을 예측하는 모델

#### Unsupervied learning : 레이블이 없는 데이터에서 구조를 식별하는 모델
- Clustering : 데이터에서 개별 그룹을 감지하고 식별하는 모델
- Dimensinality reduction : 고차원 데이터에서 저 차원 구조를 감지하고 식별하는 모델

## <font color='orange'> Basics of the API : Scikit-Learn estimator API

1. Scikit-Learn을 import 한다.
2. 모델을 선택하고 hyperparameter를 조정한다. 
3. Arrange data matrix and target vector 
4. fit()메소드를 이용해서 data를 model에 적용한다.
5. 평가한다. Apply the model to new data:
    - supervised : predict()
    - unsupervised : transform(), predict()

## <font color='blue'> 예제1. 선형회귀문제

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 10*rng.rand(50)
y = 2*x - 1 + rng.randn(50)
plt.scatter(x, y)
```

#### 1. 모델의 class를 선택한다.

```python
from sklearn.linear_model import LinearRegression
```

#### 2. 모델의 hyperparameter를 선택한다

<ol>1) class model과 instace model은 같지 않다. </ol>
<ol>2) 모델 클래스를 결정한 후에도 몇 가지 "hyperparameter"로 표현되는 몇 가지 옵션이 있음</ol>
<ol>3) 작업중인 모델 클래스에 따라 다음과 같은 하나 이상의 질문에 대답해야 할 수도 있다.
    <ol> - 오프셋에 적합합니까? (예 : y 절편)</ol> 
    <ol> - 모델을 정규화 하시겠습니까?</ol>
    <ol> - 모델 유연성을 추가하기 위해 기능을 사전 처리하고 싶습니까?</ol>
    <ol> - 모델에서 어느 정도의 정규화를 사용 하시겠습니까?</ol>
    <ol> - 몇 개의 모델 구성 요소를 사용 하시겠습니까</ol></ol>
<ol>4) 모델이 instance화 되면 우리는 오직 hyperparameter를 조정하면 된다. </ol>

```python
#절편을 hyperparameter로 선정
model = LinearRegression(fit_intercept=True)
model
```

#### 3. data를 feature matrix와 target vector로 만든다. 
- 데이터 x를 크기 [n_samples, n_features]의 행렬로 만든다

```python
x.shape
```

```python
X = x[:, np.newaxis]
X.shape
```

#### 4. Fit the model to your data
- data를 모델에 적용한다.
- .fit()을 이용

```python
model.fit(X,y)
```

- fit () 명령을 사용하면 여러 모델 종속 내부 계산이 수행되며, 계산 결과는 사용자가 탐색 할 수있는 모델 별 속성에 저장됨 
- Scikit-Learn에서 fit () 프로세스 중에 학습 된 모든 model parameter는 밑줄이 있음

```python
#기울기
model.coef_
```

```python
#y절편
model.intercept_
```

#### 5. 평가해보기 (predict labels for unknown data)
- predict() 메소드를 사용하여 unknown data에 대한 평가를 수행

```python
Xfit = np.linspace(-1, 11)
Xfit = Xfit[:, np.newaxis]
yfit = model.predict(Xfit)
```

```python
plt.scatter(x, y)
plt.plot(Xfit, yfit)
```

## <font color='blue'> 예제2. Iris Classification

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head(3)
```

```python
X_iris = iris.drop('species', axis=1)
X_iris.shape
```

```python
y_iris = iris['species']
y_iris.shape
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')

from sklearn.model_selection import train_test_split  # sklearn.cross_validation is deprecated.
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris,
                                                random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB # 1. choose model class
model = GaussianNB()                       # 2. instantiate model
model.fit(Xtrain, ytrain)                  # 3. fit model to data
y_model = model.predict(Xtest)             # 4. predict on new data
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

## <font color='blue'> 예제3. Iris dimensionality

```python
from sklearn.decomposition import PCA
model = PCA(n_components=2)
model.fit(X_iris)
X_2D = model.transform(X_iris)
```

```python
iris['PCA1'] = X_2D[:, 0]
iris['PCA2'] = X_2D[:, 1]
sns.lmplot('PCA1','PCA2',hue='species',data=iris, fit_reg=False)
```

## <font color='blue'> 예제4. Iris clustering

```python
from sklearn.mixture import GaussianMixture      # 1. Choose the model class sklearn.mixture.GaussianMixture
model = GaussianMixture(n_components=3,
            covariance_type='full')  # 2. Instantiate the model with hyperparameters
model.fit(X_iris)                    # 3. Fit to data. Notice y is not specified!
y_gmm = model.predict(X_iris)        # 4. Determine cluster labelsb
```

```python
iris['cluster'] = y_gmm
sns.lmplot('PCA1','PCA2',hue='species',col='cluster',data=iris, fit_reg=False)
```

## <font color='blue'> 응용. 손글씨 판별 Data

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.images.shape
```

```python
fig, axes = plt.subplots(10, 10, figsize=(8, 8),
                        subplot_kw={'xticks':[], 'yticks':[]},
                        gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='binary', interpolation='nearest')
    ax.text(0.05, 0.05, str(digits.target[i]),
            transform=ax.transAxes, color='green')
```

```python
X = digits.data
X.shape
```

```python
y = digits.target
y.shape
```

```python
from sklearn.manifold import Isomap
iso = Isomap(n_components=2)
iso.fit(digits.data)
data_projected = iso.transform(digits.data)
data_projected.shape
```

```python
plt.scatter(data_projected[:, 0], data_projected[:, 1], c=digits.target,
           edgecolor='none', alpha=0.5,
           cmap=plt.cm.get_cmap('Spectral', 10))
plt.colorbar(label='digit label', ticks=range(10))
plt.clim(-0.5, 9.5)
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(Xtrain, ytrain)
y_model = model.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(ytest, y_model)
sns.heatmap(mat, square=True, annot=True, cbar=False)
plt.xlabel('predicted value')
plt.ylabel('true value')
```

```python
fig, axes = plt.subplots(10, 10, figsize=(8, 8),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))

test_images = Xtest.reshape(-1, 8, 8)

for i, ax in enumerate(axes.flat):
    ax.imshow(test_images[i], cmap='binary', interpolation='nearest')
    ax.text(0.05, 0.05, str(y_model[i]),
            transform=ax.transAxes,
            color='green' if (ytest[i] == y_model[i]) else 'red')
```

## <font color='orange'> Hyperparameters and Model Validation </font>

1. choose a class of model
2. choose model hyperparameters
3. Fit the model to the training data
4. Use the model to predict labels for new data

**모델과 하이퍼 파라미터가 데이터에 잘 맞는지 검증할 방법이 필요함**

### <font color='navy'> Thinking About Model Validation </font>

- 원리: 설계한 모델과 hyperparameter에 test를 해보며 성능을 확인해 본다. 
- 즉, 모델검증: test를 통해 왜 실패하는지를 보며 보다 견고한 모델 평가를 위해 holdout set 및 cross-validation을 이용해 본다</br>

#### Holdout Sets

```python
from sklearn.model_selection import train_test_split
#split the data with 50% in each set
X1, X2, y1, y2 = train_test_split(X, y, random_state=0,
                                 train_size=0.5)

#fit the model on one set of data
model.fit(X1, y1)

#evaluate the model on the second set of data
y2_model = model.predict(X2)
accuracy_score(y2, y2_model)
```

#### Cross-Validation

```python
y2_model = model.fit(X1, y1).predict(X2)
y1_model = model.fit(X2, y2).predict(X1)
accuracy_score(y1, y1_model), accuracy_score(y2, y2_model)
```
