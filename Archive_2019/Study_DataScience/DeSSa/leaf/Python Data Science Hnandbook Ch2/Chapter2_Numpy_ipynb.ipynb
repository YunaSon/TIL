## 10.Computation on Numpy

### UFuncs
> Computation on NumPy arrays can be very fast, or it can be very slow. The key to
making it fast is to use **vectorized operations**, generally implemented through Num‐
Py’s **universal functions (ufuncs).**


````python
# The Slowness of Loops

import numpy as np
np.random.seed(0)

def compute_reciprocals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output

values = np.random.randint(1, 10,  size=5)
print(values)
compute_reciprocals(values)
````

````python
big_array = np.random.randint(1,100, size = 1000000)
%timeit compute_reciprocals(big_array)
````

### 11. Advanced Ufunc Feature
Many NumPy users make use of ufuncs without ever learning their full set of features.

````python
#Specifying output
x = np.arange(5)
y=np.empty(5)
print(x)
print(y)
print("")

np.multiply(x, 10, out=y)
print(y)
print("")

y = np.zeros(10)
print(y)
print('')

np.power(2, x, out=y[::2])
print(y)
````

````python
#Agrregates
x = np.arange(1, 6)
x_add_reduce = np.add.reduce(x)
print(x)
print(x_add_reduce)
print('')

x_multiply_reduce = np.multiply.reduce(x)
print(x_multiply_reduce)
print('')

x_add_accumulate = np.add.accumulate(x)
print(x_add_accumulate)
print('')

x_multiply_accumulate = np.multiply.accumulate(x)
print(x_multiply_accumulate)
````

### Rules of Broadcasting
* Rule 1: If the two arrays differ in their number of dimensions, the shape of the one with fewer dimensions is padded with ones on its leading (left) side.
* => 만약 차원이 다른 두 배열이면, 차원이 더 작은 쪽이 왼쪽에 채워진다.(???)
* Rule 2: If the shape of the two arrays does not match in any dimension, the array with shape equal to 1 in that dimension is stretched to match the other shape.
* => 만약 두 배열의 모양이 어느 차원이라도 맞지 않으면, 다른 쪽 모양을 맞추기 위해 그 차원의 1과 같은 배열의 모양으로 늘어납니다.(???) 
* Rule 3: If in any dimension the sizes disagree and neither is equal to 1, an error is raised.
* => 만약 어느 차원이라도 맞지 않거나 1과 같지 않으면 에러가 발생합니다.(???)

````python
#Sorting Array
import numpy as np

def selection_sort(x):
    for i in range(len(x)):
        swap = i + np.argmin(x[i:])
        (x[i], x[swap]) = (x[swap], x[i])
    return x

x = np.array([2, 1, 4, 3, 5])
selection_sort(x)
````

````python
def bogosort(x):
    while np.any(x[:-1] > x[1:]):
        np.random.shuffle(x)
    return x
x = np.array([2, 1, 4, 3, 5])
bogosort(x)
````

````python
#Example: k-Nearest Neighbors

import matplotlib.pyplot as plt
import seaborn; seaborn.set()

X = rand.rand(10, 2)
plt.scatter(X[:, 0], X[:, 1], s=100);
````

````python
dist_sq = np.sum((X[:, np.newaxis, :] - X[np.newaxis, :, :]) ** 2, axis=-1)

differences = X[:, np.newaxis, :] - X[np.newaxis, :, :]
differences.shape
````

````python
sq_differences = differences ** 2
sq_differences.shape
````

````python
dist_sq = sq_differences.sum(-1)
dist_sq.shape
````

````python
dist_sq.diagonal()
````

````python
nearest = np.argsort(dist_sq, axis=1)
print(nearest)
````

````python
K = 2
nearest_partition = np.argpartition(dist_sq, K + 1, axis=1)
````

````python
plt.scatter(X[:, 0], X[:, 1], s=100)

K = 2

for i in range(X.shape[0]):
    for j in nearest_partition[i, :K+1]:
        plt.plot(*zip(X[j], X[i]), color='black')
````
