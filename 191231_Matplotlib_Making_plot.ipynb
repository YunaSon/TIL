### plot 만들어 보기
1. dataframe용 데이터셋 : 딕셔너리의 key를 통일한다. 
2. df.iloc[0] : 인덱스 번호로 행에 접근, 
3. df.loc['sequential_mobilenetV2'] : 인덱스 이름으로 행에 접근
4. plot 만들기
   - fig = plt.figure()
   - ax1 = fig.add_subplot()
   - ax1.plot(x, y)
5. 저장 : plt.savefig()
  - 상대경로 : 현재 경로 ./
  - 상대경로 : 상위 경로 ../

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### <font color='blue'>1. Dataset 만들기 </font>

```python
# 실패 
sequential_mobilenetV2 = {'color':63.6719, 'style':52.7344, 'part':73.8281, 'season':67.5781, 'category':61.3281}
sequential_vgg = {'color':61.7188, 'style':55.0781, 'part':73.4375, 'season':69.1406, 'category':62.1094}
parallel_mobilenetV2 = {'color':61.3281, 'style':51.5625, 'part':73.4375, 'season':65.6250, 'category':53.5156}
parallel_vgg = {'color':59.7656, 'style':51.5625, 'part':71.4844, 'season':69.1406, 'category':58.9844}
parallel_mobilenetV2_style = {'color':59.7656, 'style':51.5625, 'part':71.4844, 'season':69.1406, 'category':58.9844}
parallel_vgg_style = {'color':60.9375, 'style':55.4688, 'part':73.0469, 'season':71.4844, 'category':59.7656}
```

```python
# 성공
mydict = {'color' : [63.6719, 61.7188, 61.3281, 59.7656, 59.7656, 60.9375],
'style' : [52.7344, 55.0781, 51.5625, 51.5625, 51.5625, 55.4688],
'part' : [73.8281, 73.4375, 73.4375, 71.4844, 71.4844, 73.0469],
'season' : [67.5781, 69.1406, 65.6250, 69.1406, 69.1406, 71.4844],
'category' : [61.3281, 62.1094, 53.5156, 58.9844, 58.9844, 59.7656]}

indexinfo = ['sequential_mobilenetV2', 'sequential_vgg', 'parallel_mobilenetV2', 'parallel_vgg', 'parallel_mobilenetV2_style', 'parallel_vgg_style']
```

```python
df = pd.DataFrame(mydict, index=indexinfo)
```

### <font color='blue'> 2. 인덱스 번호로 행에 접근 </font>

```python
df.iloc[0]
```

### <font color='blue'> 3. Making a Plot </font>

```python
fig = plt.figure(figsize=(18,8))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)


ax1.plot(df.iloc[0], marker=r'o', color=u'blue', linestyle='-', label='sequential_mobilenetV2')
ax1.plot(df.iloc[1], marker=r'o', color=u'red', linestyle='-', label='sequential_vgg')
ax1.plot(df.iloc[2], marker=r'o', color=u'green',linestyle='-', label='parallel_mobilenetV2')
ax1.plot(df.iloc[3], marker=r'o', color=u'black', linestyle='-', label='parallel_vgg')
ax1.plot(df.iloc[4], marker=r'o', color=u'orange', linestyle='-', label='parallel_mobilenetV2_style')
ax1.plot(df.iloc[5], marker=r'o', color=u'skyblue', linestyle='-', label='parallel_vgg_style')
ax1.set_title('kinds of classification',fontsize=20)
#ax1.ylim(50, 75)


ax2.plot(df['color'], marker=r'o', color=u'blue', linestyle='-', label='color')
ax2.plot(df['style'], marker=r'o', color=u'red', linestyle='-', label='style')
ax2.plot(df['part'], marker=r'o', color=u'green',linestyle='-', label='part')
ax2.plot(df['season'], marker=r'o', color=u'black', linestyle='-',label='season')
ax2.plot(df['category'], marker=r'o', color=u'orange', linestyle='-',label='category')
ax2.set_title('classes', fontsize=20)


ax1.set_ylabel('accuracy[%]')
ax1.set_ylim(50,75)
ax1.grid()
ax2.set_ylabel('accuracy[%]')
ax2.set_ylim(50,75)
ax2.grid()
fig.legend()
fig.show()

plt.savefig('./data/plot')
```
